```python
import pandas as pd 
import numpy as np
import seaborn as sns 
from matplotlib import pyplot as plt
```

```python
data= pd.read_csv('Student_Marks.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
df=data.copy()
```

```python
plt.plot(df)
plt.legend(df)
plt.show()
```

```python
plt.subplot(1,2,1)
sns.distplot(df['Marks'])
plt.subplot(1,2,2)
sns.distplot(df['time_study'])
plt.show()
```

```python
mean=df['Marks'].mean()
std=df['time_study'].std()
```

```python
print(mean, ",",std)
```

```python
highest_allowed=mean+3*std
lowest_allowed=mean-3*std
```

```python
print(highest_allowed,",",lowest_allowed)
```

```python
df[(df['Marks']<lowest_allowed)  | (df['time_study']>highest_allowed)]
```

## Trimming 

```python
df[(df['Marks']>lowest_allowed)  & (df['time_study']<highest_allowed)] 
```

```python
tf=pd.read_csv('bhp.csv') 
```

```python
tf
```

```python
plt.subplot(1,2,1)
sns.distplot(tf['price'])
plt.subplot(1,2,2)
sns.distplot(tf['total_sqft'])
```

```python
mean1=tf['price'].mean()
std1=tf['total_sqft'].std()
```

```python
print(mean1,',',std1)
```

```python
tf
```

```python
tf['z_score']=(tf['price']-tf['price'].mean())/tf['price'].std()
```

```python
tf.head()
```

```python
tf[(tf['z_score']< -3) | (tf['z_score']>3)]
```

```python
tf[(tf['z_score']> -3) & (tf['z_score']<3)]
```

```python
mean2=tf['price_per_sqft'].mean()
std=tf['total_sqft'].std()
```

```python
tf['z_score1']= (tf['price_per_sqft']-tf['price_per_sqft'].mean())/tf['price_per_sqft'].std()
tf.sample(10)
```

```python
tf[(tf['z_score1']<-4) | (tf['z_score1']>4)]
```

```python
tf[(tf['z_score1']>-4) & (tf['z_score1']<4)]
```

## Capping


```python
highest_allowed1=mean2+3*std
lowest_allowed1=mean2-3*std
```

```python
print(highest_allowed1,',',lowest_allowed1)
```

```python
upper_limit=mean2+3*std
lower_limit=mean2-3*std
```

```python
tf['price_per_sqft']=np.where(tf['price_per_sqft']>upper_limit,upper_limit, np.where(tf['price_per_sqft']<lower_limit,lower_limit,tf['price_per_sqft']))
```

```python
tf
```

```python
tf[(tf['price_per_sqft']<lowest_allowed1)  | (tf['price_per_sqft']>highest_allowed1)]
```
