```python
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')

imputer.fit(df7.iloc[:,-3:-2].values)

df7.iloc[:,-3:-2]=imputer.transform(df7.iloc[:,-3:-2].values)

```
